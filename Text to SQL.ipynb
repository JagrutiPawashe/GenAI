```python
from ollama import chat
```

```python
schema = """
Table: employees

Columns:
id
name
department
salary
experience
joining_date
bonus
"""
```

```python
# User question
question = input("Ask a question: ")
```

### Prompt Template

```python
# Prompt
prompt = f"""
You are an SQL expert.

Database Schema:
{schema}

Convert the user's question into MySQL query.

Return ONLY the SQL query.

Question:
{question}
"""
```

```python
# Generate SQL
response = chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)

print("\nGenerated SQL Query:\n")
print(response["message"]["content"])
```
